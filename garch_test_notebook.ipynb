# GARCH Model Testing Notebook
This notebook demonstrates how to use the `main.py` file with assertions and tests.

```python
from main import build_model, PredictIn, PredictOut
import requests
```

```python


# 1. Build model
model_shop = build_model(ticker="SHOPERSTOP.BSE", use_new_data=False)

# Assert repo is attached
assert hasattr(model_shop, 'repo'), "Model does not have repo attribute"
print("Repo attached successfully")
```

```python
# 2. Test PredictIn
pi = PredictIn(ticker="SHOPERSTOP.BSE", n_days=5)
print(pi)

# 3. Test PredictOut
po = PredictOut(ticker="SHOPERSTOP.BSE", n_days=5, success=True, forecast={}, message="success")
print(po)

assert isinstance(po.dict(), dict), "PredictOut is not a dictionary-like output"
print("PredictOut validation passed")
```

```python
# 4. Test FastAPI endpoint (if running)
url = "http://localhost:8010/predict"
json_data = {"ticker":"SHOPERSTOP.BSE","n_days":3}

try:
    response = requests.post(url=url, json=json_data)
    submission = response.json()
    print(submission)
    assert 'forecast' in submission, "No forecast in response"
except Exception as e:
    print("FastAPI not running or error:", e)
```
